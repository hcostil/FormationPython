```python
import numpy as np
import matplotlib.pyplot as plt
```

### Visualisation de données avec Matplotlib
Ce notebook contient plusieurs exemples d'utilisation de **Matplotlib** pour générer différents types de graphiques : courbe, histogramme, diagramme en barres et nuage de points.

```python
# Génération de données factices
x = np.linspace(0, 10, 100)
y = np.sin(x)
data = np.random.randn(1000)
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]
```

```python
# Création d'une figure avec plusieurs sous-graphiques
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# 1. Tracé d'une courbe
axs[0, 0].plot(x, y, label='Sinus', color='b', linestyle='--')
axs[0, 0].set_title("Courbe sinus")
axs[0, 0].set_xlabel("X")
axs[0, 0].set_ylabel("Sin(X)")
axs[0, 0].legend()

# 2. Histogramme
axs[0, 1].hist(data, bins=30, color='g', alpha=0.7, edgecolor='black')
axs[0, 1].set_title("Histogramme des valeurs aléatoires")

# 3. Diagramme en barres
axs[1, 0].bar(categories, values, color='purple')
axs[1, 0].set_title("Diagramme en barres")

# 4. Nuage de points
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
axs[1, 1].scatter(x_scatter, y_scatter, color='orange', alpha=0.7)
axs[1, 1].set_title("Nuage de points")

# Ajustement automatique des espaces entre sous-graphiques
plt.tight_layout()

# Affichage des graphiques
plt.show()
```
