
# Exemples d'utilisation de Matplotlib

Ce notebook présente différents exemples d'utilisation de la bibliothèque **Matplotlib** pour la visualisation de données en Python.


```python

# Installation (si nécessaire)
# !pip install matplotlib

# Importation de la bibliothèque
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Exemple de graphique en ligne

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y, marker='o', linestyle='--', color='r', label="Évolution")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Exemple de graphique en ligne")
plt.legend()
plt.grid(True)
plt.show()
```

```python

# Exemple d'histogramme

data = np.random.randn(1000)

plt.hist(data, bins=30, color="g", alpha=0.7, edgecolor="black")
plt.title("Histogramme")
plt.xlabel("Valeurs")
plt.ylabel("Fréquence")
plt.show()
```

```python

# Exemple de diagramme en barres

categories = ["A", "B", "C", "D"]
values = [5, 7, 3, 8]

plt.bar(categories, values, color="purple")
plt.title("Diagramme en barres")
plt.xlabel("Catégories")
plt.ylabel("Valeurs")
plt.show()
```

```python

# Exemple de nuage de points

x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color="orange", alpha=0.7)
plt.title("Nuage de points")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python

# Exemple de graphique en secteurs (camembert)

labels = ["Python", "Java", "C++", "JavaScript"]
sizes = [40, 25, 20, 15]

plt.pie(sizes, labels=labels, autopct="%1.1f%%", colors=["blue", "red", "green", "orange"])
plt.title("Répartition des langages")
plt.show()
```

```python

# Exemple d'affichage de plusieurs graphiques avec subplot

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Graphique en ligne
axs[0, 0].plot([1, 2, 3, 4], [10, 20, 30, 40], color="blue")
axs[0, 0].set_title("Ligne")

# Histogramme
axs[0, 1].hist(np.random.randn(1000), bins=20, color="green")
axs[0, 1].set_title("Histogramme")

# Diagramme en barres
axs[1, 0].bar(["A", "B", "C"], [10, 5, 7], color="red")
axs[1, 0].set_title("Barres")

# Nuage de points
axs[1, 1].scatter(np.random.rand(50), np.random.rand(50), color="purple")
axs[1, 1].set_title("Nuage de points")

plt.tight_layout()
plt.show()
```

```python

# Exemple de personnalisation avancée avec annotations et style

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label="sin(x)", color="blue", linewidth=2)
plt.axhline(0, color="gray", linestyle="--")
plt.axvline(np.pi, color="red", linestyle="--", label="x=π")

# Annotation
plt.annotate("Point clé", xy=(np.pi, 0), xytext=(np.pi, 0.5),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Exemple de personnalisation avancée")
plt.legend()
plt.grid(True)
plt.show()
```
