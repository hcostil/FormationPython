# üéÆ Analyse des performances des joueurs avec Seaborn
## üìå Objectif
Ce notebook permet d'explorer un jeu de donn√©es contenant les performances des joueurs dans un jeu vid√©o.
Nous utiliserons **Seaborn** pour analyser visuellement ces donn√©es.

## üìä Donn√©es
Le dataset contient les informations suivantes :
- **player_id** : Identifiant unique du joueur.
- **score** : Score obtenu par le joueur.
- **time_played** : Temps pass√© sur le jeu (en heures).
- **level** : Niveau atteint par le joueur.
- **game_mode** : Mode de jeu choisi (Solo, Multi, Coop).
- **win_rate** : Pourcentage de victoires.


```python
# üìå Importation des biblioth√®ques
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# üìå Chargement des donn√©es (Simulation d'un dataset)
import numpy as np
np.random.seed(42)

n = 500  # Nombre de joueurs
data = {
    'player_id': range(1, n+1),
    'score': np.random.randint(1000, 5000, n),
    'time_played': np.random.uniform(5, 50, n),
    'level': np.random.randint(1, 20, n),
    'game_mode': np.random.choice(['Solo', 'Multi', 'Coop'], n),
    'win_rate': np.random.uniform(20, 80, n)
}

df = pd.DataFrame(data)

# üìå Aper√ßu des donn√©es
df.head()
```

## üìà 1Ô∏è‚É£ Distribution des scores
Cr√©ons un **histogramme** des scores des joueurs pour observer leur r√©partition.

```python
# üìä Histogramme de la distribution des scores
plt.figure(figsize=(10,5))
sns.histplot(df['score'], bins=30, kde=True, color='royalblue')
plt.title('Distribution des scores des joueurs')
plt.xlabel('Score')
plt.ylabel('Nombre de joueurs')
plt.show()
```

## üîπ 2Ô∏è‚É£ Relation entre le score et le temps de jeu
Nous allons cr√©er un **nuage de points** pour observer s'il existe une corr√©lation entre le score et le temps jou√©.

```python
# üìä Scatter plot entre score et temps de jeu
plt.figure(figsize=(10,5))
sns.scatterplot(data=df, x='time_played', y='score', hue='game_mode', alpha=0.7)
plt.title('Relation entre le score et le temps de jeu')
plt.xlabel('Temps jou√© (heures)')
plt.ylabel('Score')
plt.show()
```

## üî∏ 3Ô∏è‚É£ R√©partition des scores selon le niveau
Utilisons un **boxplot** pour analyser les scores selon les niveaux atteints.

```python
# üìä Boxplot des scores par niveau
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x='level', y='score', palette='coolwarm', hue='level', legend=False)
plt.title('R√©partition des scores selon le niveau atteint')
plt.xlabel('Niveau')
plt.ylabel('Score')
plt.show()
```

## üî∫ 4Ô∏è‚É£ Scores moyens par mode de jeu
Utilisons un **diagramme en barres** pour afficher la moyenne des scores selon le mode de jeu.

```python
# üìä Barplot des scores moyens par mode de jeu
plt.figure(figsize=(8,5))
sns.barplot(data=df, x='game_mode', y='score', errorbar=None, palette='Set2', hue='game_mode', legend=False)
plt.title('Score moyen par mode de jeu')
plt.xlabel('Mode de jeu')
plt.ylabel('Score moyen')
plt.show()
```

## üü¢ 5Ô∏è‚É£ Matrice de corr√©lation
Observons les corr√©lations entre les diff√©rentes variables √† l'aide d'une **heatmap**.

```python
# üìä Heatmap des corr√©lations (correction)
plt.figure(figsize=(8,6))
corr = df.drop(columns=['game_mode']).corr()  # Suppression des colonnes non num√©riques
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Corr√©lation entre les variables')
plt.show()
```
