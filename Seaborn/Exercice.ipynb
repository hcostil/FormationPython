# üìä Exercice Seaborn : Analyse des performances des joueurs
## Contexte
Nous travaillons pour un studio de jeu vid√©o et nous voulons analyser
les performances des joueurs √† travers diff√©rentes m√©triques.
Nous utiliserons **Seaborn** pour visualiser ces donn√©es et comprendre les tendances.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Chargement des donn√©es fictives
data = {
    "Joueur": ["Alice", "Bob", "Charlie", "David", "Eve", "Frank", "Grace", "Hugo", "Ivy", "Jack"],
    "Score": [1500, 2300, 1800, 1200, 2500, 3000, 2700, 2100, 1900, 2200],
    "Temps_Jeu": [35, 50, 40, 30, 55, 65, 60, 45, 38, 48],
    "Niveau": [5, 7, 6, 4, 8, 9, 8, 6, 5, 7],
    "Type": ["Attaquant", "D√©fenseur", "Attaquant", "D√©fenseur", "Attaquant", "D√©fenseur", "Attaquant", "D√©fenseur", "Attaquant", "D√©fenseur"]
}

df = pd.DataFrame(data)
df.head()  # Afficher les premi√®res lignes des donn√©es
```

## üìå Exercice 1 : Corr√©lation entre Score et Temps de jeu
Cr√©er un diagramme en nuage de points pour visualiser la corr√©lation entre Score et Temps de jeu.

```python
sns.___(data=df, x="Temps_Jeu", y="Score", hue="Type", style="Type")
plt.title("Corr√©lation entre le Temps de Jeu et le Score")
plt.show()
```

## üìå Exercice 2 : Analyse de la distribution des scores
Afficher un **boxplot** pour visualiser la distribution des scores par type de joueur.

```python
sns.___(data=df, x="Type", y="Score")
plt.title("Distribution des Scores par Type de Joueur")
plt.show()
```

## üìå Exercice 3 : R√©partition des niveaux des joueurs
Afficher un **histogramme** pour analyser la r√©partition des niveaux des joueurs.

```python
sns.___(df["Niveau"], bins=5, kde=True)
plt.title("Distribution des Niveaux des Joueurs")
plt.show()
```

## üìå Exercice 4 : Heatmap des corr√©lations
Cr√©er une **heatmap** pour visualiser les corr√©lations entre les diff√©rentes m√©triques des joueurs.

```python
corr = df.drop(columns=['Joueur', 'Type']).corr()  # Suppression des colonnes non num√©riques
sns.___(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Corr√©lation entre les m√©triques des joueurs")
plt.show()
```
