```python

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Chargement du dataset "iris"
iris = sns.load_dataset("iris")

# Affichage des premières lignes
display(iris.head())
```

```python

# 1. Nuage de points (scatterplot)
plt.figure(figsize=(8, 5))
sns.scatterplot(data=iris, x="petal_length", y="petal_width", hue="species", style="species")
plt.title("Relation entre la longueur et la largeur des pétales")
plt.show()
```

```python

# 2. Histogramme et distribution
plt.figure(figsize=(8, 5))
sns.histplot(data=iris, x="petal_length", hue="species", kde=True, bins=20)
plt.title("Distribution des longueurs de pétales par espèce")
plt.show()
```

```python

# 3. Boîte à moustaches (boxplot)
plt.figure(figsize=(8, 5))
sns.boxplot(data=iris, x="species", y="sepal_length")
plt.title("Distribution des longueurs de sépales par espèce")
plt.show()
```

```python

# 4. Diagramme en barres (barplot)
plt.figure(figsize=(8, 5))
sns.barplot(data=iris, x="species", y="sepal_length", errorbar=None)
plt.title("Moyenne des longueurs de sépales par espèce")
plt.show()
```

```python

# 5. Carte de chaleur (heatmap)
corr = iris.drop(columns=["species"]).corr()
plt.figure(figsize=(8, 5))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Corrélation entre les mesures des fleurs")
plt.show()
```

```python

# 6. Pairplot pour visualiser toutes les relations
sns.pairplot(iris, hue="species")
plt.show()
```

```python

# 7. Personnalisation avec styles et palettes
sns.set_style("darkgrid")
sns.set_palette("pastel")

plt.figure(figsize=(8, 5))
sns.histplot(data=iris, x="petal_length", hue="species", kde=True, bins=15)
plt.axvline(iris["petal_length"].mean(), color="red", linestyle="--", label="Moyenne")
plt.legend()
plt.show()
```
