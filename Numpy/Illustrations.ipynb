# Exercice NumPy : Manipulation avancée des tableaux NumPy


## Objectif
Cet exercice a pour but d'explorer différentes fonctionnalités de **NumPy**, une bibliothèque puissante pour le calcul scientifique en Python.
Vous allez manipuler des **tableaux**, réaliser des **opérations avancées**, explorer le **broadcasting**, et appliquer NumPy à des contextes concrets.
    

## 1. Création et manipulation de tableaux NumPy

```python

import numpy as np

# Création d'un tableau 1D
a = np.array([1, 2, 3, 4, 5])
print("Tableau 1D:", a)

# Création d'une matrice 2D
b = np.array([[1, 2, 3], [4, 5, 6]])
print("Matrice 2D:\n", b)

# Création d'un tableau de zéros et de uns
zeros = np.zeros((3, 3))
ones = np.ones((3, 3))
print("Matrice de zéros:\n", zeros)
print("Matrice de uns:\n", ones)

# Création d'un tableau avec valeurs aléatoires
random_array = np.random.rand(3, 3)
print("Tableau aléatoire:\n", random_array)
    
```

## 2. Opérations mathématiques et statistiques

```python

# Définition d'un tableau
x = np.array([10, 20, 30, 40, 50])

# Opérations mathématiques
print("Addition:", x + 5)
print("Multiplication:", x * 2)
print("Exponentielle:", np.exp(x))

# Statistiques
print("Moyenne:", np.mean(x))
print("Médiane:", np.median(x))
print("Écart-type:", np.std(x))
    
```

## 3. Indexation, slicing et broadcasting

```python

# Création d'un tableau 2D
arr = np.array([[10, 20, 30], [40, 50, 60], [70, 80, 90]])
print("Tableau:\n", arr)

# Indexation
print("Élément (1,2) :", arr[1, 2])

# Slicing
print("Ligne 1 :", arr[1, :])
print("Colonne 2 :", arr[:, 2])

# Broadcasting
arr[:, 1] = arr[:, 1] * 2  # Multiplie la colonne 1 par 2
print("Après broadcasting:\n", arr)
    
```

## 4. Génération de données aléatoires pour simulations

```python

# Génération de nombres aléatoires suivant une distribution normale
random_values = np.random.normal(loc=0, scale=1, size=1000)
print("Quelques valeurs générées :", random_values[:10])

# Calcul de la moyenne et de l'écart-type
print("Moyenne:", np.mean(random_values))
print("Écart-type:", np.std(random_values))
    
```

## 5. Application : Traitement d'image avec NumPy

```python

import matplotlib.pyplot as plt

# Création d'une image 100x100 en niveaux de gris
image = np.random.rand(100, 100)

# Affichage de l'image
plt.imshow(image, cmap='gray')
plt.colorbar()
plt.title("Image générée avec NumPy")
plt.show()
    
```

## 6. Produit matriciel et application en algèbre linéaire

```python

# Création de deux matrices
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# Produit matriciel
C = np.dot(A, B)
print("Produit matriciel:\n", C)

# Calcul des valeurs propres
valeurs_propres, vecteurs_propres = np.linalg.eig(A)
print("Valeurs propres:", valeurs_propres)
print("Vecteurs propres:\n", vecteurs_propres)
    
```
